```python
import pandas as pd

# Load the data
data = pd.read_csv('/Users/won-yooseung/Desktop/16_KRX_주식_투자_알고리즘_경진대회/train_new.csv')
data.head()
```

```python
# Convert the '일자' column to datetime format
data['일자'] = pd.to_datetime(data['일자'], format='%Y%m%d')

data.head()
```

```python
# Sort the data by '종목코드' and '일자'
data = data.sort_values(['종목코드', '일자'])

# Calculate the return for each stock over the next 15 trading days
data['수익률'] = data.groupby('종목코드')['종가'].pct_change(periods=-15)

data.head(20)
```

```python
# Rank the returns for each day
data['수익률 순위'] = data.groupby('일자')['수익률'].rank(ascending=False)

data.head(20)
```

```python
# Calculate the average rank for each stock
average_rank = data.groupby('종목코드')['수익률 순위'].mean().reset_index()
average_rank.columns = ['종목코드', '평균 순위']
average_rank.head()
```

```python
# Calculate the average return for each stock
average_return = data.groupby('종목코드')['수익률'].mean().reset_index()
average_return.columns = ['종목코드', '평균 수익률']

# Rank the average returns
average_return['수익률 순위'] = average_return['평균 수익률'].rank(ascending=False)

average_return.head()
```

```python
# Merge the two dataframes
merged = pd.merge(average_rank, average_return, on='종목코드')

# Calculate the final rank
merged['최종 순위'] = merged[['평균 순위', '수익률 순위']].mean(axis=1)

merged.head()
```

```python
# Sort the dataframe by the final rank
sorted_df = merged.sort_values('최종 순위')

# Reset the index to get the final rank from 1 to 2000
sorted_df = sorted_df.reset_index(drop=True)
sorted_df['최종 순위'] = sorted_df.index + 1

sorted_df.head()
```

```python
# Save the dataframe to a csv file
final_df = sorted_df[['종목코드', '최종 순위']]
final_df.to_csv('/Users/won-yooseung/Desktop/16_KRX_주식_투자_알고리즘_경진대회/final_rank_new.csv', index=False)
```
